```python
# --- Check Class Distribution ---
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, palette='coolwarm')
plt.title('Class Distribution (Fraud vs Legit)')
plt.xlabel('Class (0 = Legit, 1 = Fraud)')
plt.ylabel('Count')
plt.show()
```

Insight: Fraudulent transactions are typically much fewer than legitimate ones.

```python
# Filter transactions with amounts less than 7500
filtered_df = df[df['Amount'] < 7500]

# Custom colors for 'Legit' and 'Fraud'
custom_palette = {0: "blue", 1: "red"}  # 0 = Legit (blue), 1 = Fraud (red)

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=filtered_df, x='Time', y='Amount', hue='Class', palette=custom_palette, alpha=0.6)
plt.title('Transaction Amount Over Time (Amount < 7500)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amount ($)')
plt.legend(title='Class', labels=['Legit', 'Fraud'], loc='upper right')
plt.show()
```

 Insight: Examine periods with more fraudulent transactions or anomalies in transaction amounts.


```python
# Check if 'Amount' and 'Class' columns exist
if 'Amount' in df.columns and 'Class' in df.columns:
    plt.figure(figsize=(10, 6))

    # Ensure 'Class' is treated as categorical if needed
    if df['Class'].dtype != 'category':
        df['Class'] = df['Class'].astype('category')

    # Plot
    sns.histplot(data=df, x='Amount', bins=30, kde=True, hue='Class', palette='coolwarm', alpha=0.5)
    plt.title('Transaction Amount Distribution by Class')
    plt.xlabel('Amount ($)')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Error: 'Amount' or 'Class' column not found in the dataset.")
```

```python
# --- Distribution of Transaction Time ---
plt.figure(figsize=(8, 5))
sns.histplot(df['Time'], bins=10, kde=True, color='blue')
plt.title('Distribution of Transaction Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency')
plt.show()
```

```python
# --- Distribution of Transaction Amount ---
plt.figure(figsize=(8, 5))
sns.histplot(df['Amount'], bins=10, kde=True, color='green')
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

 Insight: Fraudulent transactions might be concentrated in specific amount ranges.


```python
#--- Correlation Heatmap ---
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=False, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Feature Correlation Heatmap')
plt.show()
```

Insight: Fraudulent transactions often deviate significantly in terms of amount compared to legitimate transactions.


```python
# --- Pairplot for Important Features ---
features_to_plot = ['V1', 'V2', 'V3', 'V4','Class']
sns.pairplot(df[features_to_plot], hue='Class', palette='coolwarm', diag_kind='kde', plot_kws={'alpha': 0.5})
plt.show()
```

Insight: Detect time periods when fraudulent activities spike.

```python
# --- Distribution of Features ---
plt.figure(figsize=(20, 10))
for i, col in enumerate(df.columns[1:29], 1):  # Assuming V1-V28 are feature columns
    plt.subplot(7, 4, i)
    sns.histplot(data=df, x=col, hue='Class', kde=True, bins=30, palette='coolwarm', alpha=0.6)
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```
